# chuyển số thành hình

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Đọc file csv
df = pd.read_csv('../data/raw/ecg.csv', header=None)
label_col = df.columns[-1]

# ======================================================
# BƯỚC QUAN TRỌNG: TÌM ĐỈNH VÀ ĐÁY CỦA TOÀN BỘ DỮ LIỆU
# ======================================================
# Lấy toàn bộ dữ liệu sóng (bỏ cột nhãn)
all_signals = df.iloc[:, :-1].values
global_min = all_signals.min() - 0.5 # Trừ hao thêm 0.5 cho nó đỡ lẹm viền
global_max = all_signals.max() + 0.5 # Cộng thêm 0.5 cho có khoảng trống

print(f"Chốt sổ trục Y: Min = {global_min:.2f}, Max = {global_max:.2f}")

# Tạo thư mục
os.makedirs('../data/preprocessed/image/normal', exist_ok=True)
os.makedirs('../data/preprocessed/image/abnormal', exist_ok=True)

print("Convert sóng ECG thành ảnh đang diễn ra... Xin đợi tí nhé!")

# Chuẩn bị 1 cái khung ảnh duy nhất (xài chung cho lẹ, đỡ tràn RAM)
# Chỉnh figsize=(2.24, 2.24) và dpi=100 -> Ảnh xuất ra sẽ đúng chuẩn 224x224 pixel
fig, ax = plt.subplots(figsize=(2.24, 2.24), dpi=100)

for idx, row in df.iterrows():
    label = row[label_col]
    data_points = row.values[:-1]

    # Xóa nội dung của cái khung ảnh cũ trước khi vẽ nét mới
    ax.clear()
    
    # Vẽ sóng
    ax.plot(data_points, color='black', linewidth=1.5)
    
    # KHÓA TRỤC Y LẠI (QUAN TRỌNG VÃI L)
    ax.set_ylim(global_min, global_max)
    
    # Tắt trục tọa độ
    ax.axis('off')

    # Định tuyến lưu file
    if label == 1: 
        filename = f'../data/preprocessed/image/normal/ecg_{idx}.jpg' # Đổi sang jpg cho nhẹ
    else:          
        filename = f'../data/preprocessed/image/abnormal/ecg_{idx}.jpg'

    # Lưu ảnh, bỏ cái viền trắng thừa thãi đi
    fig.savefig(filename, bbox_inches='tight', pad_inches=0, format='jpg')

plt.close(fig) # Xong xuôi thì đóng lại
print("complete! Tất cả sóng ECG đã được chuyển thành ảnh và lưu vào thư mục tương ứng rồi nhé!")
```
